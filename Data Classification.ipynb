```python
# duplicate data
data_1 = data.copy()
# Remove major
data_1 = data_1.drop(['major', 'country', 'uniquenetworklocation', 'source'],axis=1)
# Remove time and position
time = [i for i in data_1.iloc[:,0:126] if  'E' in i]
position = [i for i in data_1.iloc[:,0:126] if  'I' in i]
data_1 = data_1.drop(position,axis=1)
data_1 = data_1.drop(time,axis=1)
# remove introelapse, testelapse, surveyelapse
data_1 = data_1.drop(data_1.iloc[:,42:45],axis=1)
data_2 = data_1.copy()
# Remove VCL
# data_2 = data_2.drop(data_2.iloc[:,52:68],axis=1)
# data_2 = data_2.replace(to_replace=0,value=3)
# Rename Ten Item Personlaity code
data_2 = data_2.rename(columns={'TIPI1':'Extraverted-enthusiastic','TIPI2':'Critical-quarrelsome',
                            'TIPI3':'Dependable-self_disciplined','TIPI4':'Anxious-easily upset',
                            'TIPI5':'Open to new experiences-complex','TIPI6':'Reserved-quiet',
                            'TIPI7':'Sympathetic-warm','TIPI8':'Disorganized-careless','TIPI9':'Calm-emotionally_stable',
                            'TIPI10':'Conventional-uncreative'})
#print('Shape',data_1.shape)
#print('Attributes',data_1.columns)
print('Shape',data_2.shape)
print('Attributes',data_2.columns)
```

```python
# Grouping age
def condition(x):
    if x <= 10:
        return 0 #'Under Age'
    if  10 <= x <= 16:
        return 1 #'Children'
    if 17 <= x <= 21:
        return 2 #'Teen'
    if 21 <= x <= 35:
        return 3 #'Adults'
    if 36 <= x <= 48:
        return 4 #'Elder Adults'
    if x >= 49:
        return 5 #'Older People'

data_2['AgeGroups'] = data_2['age'].apply(condition)
data_2.head()
```

```python
# Filter DAS
data_das = data_2.iloc[:,42:]
#print('Attributes',data_das.columns)
data_3=data_2.filter(regex='Q\d{1,2}A')
#print('Attributes',data_3.columns)
data_3.head()
```

```python
# Categorized DAS
def sub(data_3):
    return data_3.subtract(1,axis=1)
data_3 = sub(data_3)
das_keys = {'Depression': [3, 5, 10, 13, 16, 17, 21, 24, 26, 31, 34, 37, 38, 42],
             'Anxiety': [2, 4, 7, 9, 15, 19, 20, 23, 25, 28, 30, 36, 40, 41],
             'Stress': [1, 6, 8, 11, 12, 14, 18, 22, 27, 29, 32, 33, 35, 39]}
d = []
for i in das_keys["Depression"]:
    d.append('Q'+str(i)+'A')
s = []
for i in das_keys["Stress"]:
    s.append('Q'+str(i)+'A')
a = []
for i in das_keys["Anxiety"]:
    a.append('Q'+str(i)+'A')
depression = data_3.filter(d)
stress = data_3.filter(s)
anxiety = data_3.filter(a)
```

```python
# scoring system
def scores(source):
    col=list(source)
    source['Total_Count']=source[col].sum(axis=1)
    return source
depression=scores(depression)
stress=scores(stress)
anxiety=scores(anxiety)

display(depression.head(), stress.head(), anxiety.head())
```

```python
from sklearn.ensemble import RandomForestClassifier

#using random forest for calculate features importance, returnung importances of each feature
def get_feature_importances(data_2, train_data, label_data):
  classifier = RandomForestClassifier(n_estimators = 2000, n_jobs=-1)
  classifier.fit(train_data, label_data)
  
  result = dict()
  result['features'] = data_2.drop(columns='Condition').columns #extract label for each feature
  result['importances'] = classifier.feature_importances_ #get feature importance for each feature
  result['index']= np.argsort( result['importances'])[::-1] #sort feature importances from high to low
  result['classifier'] = classifier
  
  return result
```

```python
#return dataframe of rank of feature importances
def show_importance_features(feature_importance, title):
  return pd.DataFrame(data = feature_importance['importances'][feature_importance['index']], 
               index = feature_importance['features'][feature_importance['index']], 
               columns = ['Importances {}'.format(title)])
```

```python
#show plot importances of each features
def show_importance_feature_plot(feature_importance, train_data, title):
  plt.figure(figsize = (25, 5))
  plt.title('Feature Importances {}'.format(title))
  plt.bar(range(train_data.shape[1]), feature_importance['importances'][feature_importance['index']], color=['lightgreen'], align='center')#height of each bar based on its importance
  plt.xticks(range(train_data.shape[1]), feature_importance['features'][feature_importance['index']], rotation=90)#label for each bar 'Features name'
  plt.show()
```

```python
from sklearn.feature_selection import SelectFromModel

#get selected feature from the importances using selectFromModel
def get_importance_features(classifier, train_data, label_data, data_2):
  select_feature = SelectFromModel(classifier)
  select_feature.fit(train_data, label_data)
  selected_features = data_2.drop(columns='Condition').columns[select_feature.get_support()]
  return selected_features
```
